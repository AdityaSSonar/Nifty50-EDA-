Converting Object type to Float 
(Open, High, Low,LTP)

```python
for i in df.select_dtypes(include = 'object').columns[1:]:
    df[i] = df[i].str.replace(",","")
    df[i] = df[i].astype(float)
df
```

# BARPLOT  VISUALIZATION

```python
plt.figure(figsize=(10,5))
plt.xticks(rotation=45)
sns.barplot(data = df, x = 'Symbol', y = '30 d % chng')
plt.show
```

INTERPRETATION :-
Many companies often have negative returns within a month of investment, so it's better to consider longer investment periods.

```python
plt.figure(figsize=(14,7))
plt.xticks(rotation=45)
sns.barplot(data = df, x = 'Symbol', y = '365 d % chng')
plt.show
```

INTERPRETATION :-
Observe that many companies have delivered returns of at least 10% for investors who remained invested in them for a year or longer.

```python
plt.figure(figsize=(15,8))
plt.xticks(rotation=45)
sns.barplot(data = df, x = 'Symbol', y = 'Chng')
plt.show
```

INTERPRETATION :-
Most of the Stocks gave Negative ↓ returns .

```python
df[(df['Chng']>0)]
plt.figure(figsize=(10,7))
plt.xticks(rotation=45)
sns.barplot(data = df[(df['Chng']>0)], x = 'Symbol', y = 'Chng')
plt.show
```

INTERPRETATION :-
Only 4 stocks had given Positive ↑ Returns.

```python
change_in_year=df.sort_values('365 d % chng')
change_in_year
```

```python
plt.figure(figsize=(10,7))
plt.xticks(rotation=45)
sns.barplot(data = change_in_year, x = 'Symbol', y = '365 d % chng')
plt.title("Yearly Growth(Sorted)")  
plt.show
```

INTERPRETATION :-
Tata Motors had an outstanding year with returns soaring over 167.95%, while Hero MotoCorp had a tougher time, showing returns lower than -16.02% for the same year.

```python
nifty_50=df.iloc[:,5:]
nifty_50_corr=nifty_50.corr()
plt.figure(figsize=(10, 10))
sns.heatmap(nifty_50_corr,annot=True)
plt.title('heatmapp')
```

INTERPRETATION :-
The current percent change is positively related to the 30-day percent change(0.56) but negatively related to the 365-day percent change(-0.39). This implies that short-term changes drive prices, while longer-term changes have less impact.
